```python
# importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://bookpage.com/reviews?book_genre=fiction&page=1")
page
```

```python
soup= BeautifulSoup(page.content)
soup
```

```python
soup1 = BeautifulSoup(page.content, 'html.parser')
```

```python
soup1
```

```python

soup1.title.text   
```

```python
all_book = soup.findAll('div',{'class':'row-fluid article-row'})
```

```python
all_book
```

```python
first_book =all_book[0]
```

```python
first_book
```

```python
first_book.find('h4',{'class':'italic'}).a.text
```

```python
first_book.find('p',{'class':'sans bold'}).text.replace('\n','')
```

```python
first_book.find('p',{'class':'genre-links hidden-phone'}).text.replace('\n','')
```

```python
Name=[]
Author=[]
Genre=[]
for i in all_book:
    
    Name.append(i.find('h4',{'class':'italic'}).a.text)
    Author.append(i.find('p',{'class':'sans bold'}).text.replace('\n',''))
    Genre.append(i.find('p',{'class':'genre-links hidden-phone'}).text.replace('\n',''))
    break
    
```

```python
data= list(zip(Name,Author,Genre))
```

```python
import pandas as pd
df =pd.DataFrame(data,columns=['Name',"Author","Genre"])
```
