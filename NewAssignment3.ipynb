#### Q1- Write a python program which searches all the product under a particular product vertical from www.amazon.in. The product verticals to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.
#### Q2- In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
driver = webdriver.Chrome("chromedriver.exe")     
url = "https://www.amazon.in/"
driver.get(url)

time.sleep(2)
product = input('Enter the product name : ')
```

```python
search = driver.find_element_by_id("twotabsearchtextbox")    
search.clear()                                              
search.send_keys(product)                                  
searchbutton = driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]/span/input')     
searchbutton.click()                                                                
```

```python
product_urls = []

for page in range(0,3):
    url=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')      
    for i in url:
        product_urls.append(i.get_attribute("href"))                        
    nxt_button=driver.find_element_by_xpath("//li[@class='a-normal'][1]/a")    
    nxt_button.click()
    time.sleep(2)
len(product_urls)
```

```python

Brand = []   
Name = []
Rating = []
NoOfRatings = []
Price = []
Return = []
deliveryexpected = []
Availability = [] 
OtherDetails = []

for url in product_urls:
    driver.get(url)
    try:
        brand = driver.find_element_by_xpath('//a[@id="bylineInfo"]')      
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')    

    try:
        name = driver.find_element_by_id('productTitle')      
        Name.append(name.text)
    except NoSuchElementException:
        Name.append('-')

    try:
        rating = driver.find_element_by_xpath('//span[@class="a-size-base a-nowrap"]//span')  
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append('-')


    try:
        noRating = driver.find_element_by_id('acrCustomerReviewText')  
        NoOfRatings.append(noRating.text)
    except NoSuchElementException:
        NoOfRatings.append('-')


    try:
        price = driver.find_element_by_xpath('//td[@class="a-span12"]')  
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')


    try:
        preturn = driver.find_element_by_xpath('//a[@class="a-size-small a-link-normal a-text-normal"]')  
        Return.append(preturn.text)
    except NoSuchElementException:
        Return.append('-')

        
    try:
        delivery= driver.find_element_by_xpath('//span[@class="a-color-error"]')  
        deliveryexpected.append(delivery.text)
    except NoSuchElementException:
         deliveryexpected.append('-')

    try:
        Pavailibity= driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')  
        Availability.append(Pavailibity.text)
    except NoSuchElementException:
        Availability.append('-')
 
    try:
        Details= driver.find_element_by_xpath('//ul[@class="a-unordered-list a-vertical a-spacing-mini"]')  
        OtherDetails.append(Details.text)
    except NoSuchElementException:
        OtherDetails.append('-')

time.sleep(3)    
  
```

```python
prodt=pd.DataFrame({})
prodt['Brand']=Brand
prodt['Name']=Name
prodt['Rating']=Rating
prodt['No of ratings']=NoOfRatings
prodt['Price']=Price
prodt['Return/Exchange']=Return
prodt['Delivery Expected']=deliveryexpected
prodt['Availability']=Availability
prodt['Other Details']=OtherDetails
prodt['URL']=product_urls
prodt
```

```python
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim

longitude = []
latitude = []
   

def findGeocode(city):

    try:
        geolocator = Nominatim(user_agent="your_app_name")
          
        return geolocator.geocode(city)
      
    except GeocoderTimedOut:
          
        return findGeocode(city)    

for i in (df["City"]):
      
    if findGeocode(i) != None:
           
        loc = findGeocode(i)
          
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)
        
    else:
        latitude.append(np.nan)
        longitude.append(np.nan)
```

```python

df["Longitude"] = longitude
df["Latitude"] = latitude
  
df
```

#### 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
data=[]

with Chrome(executable_path='chromedriver.exe') as driver:
    wait = WebDriverWait(driver,15)
    driver.get("https://www.youtube.com/watch?v=kuhhT_cBtFU&t=2s")

    for item in range(5): 
        wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
        time.sleep(15)

    for comment in wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#content"))):
        data.append(comment.text)   
 #note range will set for 500 comment 250
```
