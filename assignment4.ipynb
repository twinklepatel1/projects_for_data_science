```python
driver = webdriver.Chrome("chromedriver.exe")     
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)

```

```python
rank=[]
name_titile=[]
artistName=[]
uploadDate1=[]
view1=[]
rank1=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]")
rank1
for i in rank1:      
    rank.append(i.text.replace(".",""))
    
name=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]")
name
for i in name:      
    name_titile.append(i.text)  
    
artist=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]")
artist
for i in artist:      
    artistName.append(i.text)  

uploadDate=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]")
uploadDate
for i in uploadDate:      
    uploadDate1.append(i.text) 
    
view=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]")
view
for i in view:      
    view1.append(i.text)  
```

```python
DT=pd.DataFrame({})


DT['rank']=rank1
DT['Name Titile']=name_titile[:30]
DT['Artist Name']=artistName[:30]
DT['upload date']=uploadDate1[:30]
DT['View (in bilion)']=view1[:30]

DT
```
